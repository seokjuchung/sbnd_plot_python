```python
import matplotlib.pyplot as plt
import numpy as np
import sbnd_style as sbnd_style # Applies the style automatically

x=np.arange(0,100,1)
y=2*x
z=3*x
a=4*x
b=5*x
c=6*x

fig, ax = plt.subplots()
ax.plot(x,y)
ax.plot(x,z)
ax.plot(x,a)
ax.plot(x,b)
ax.plot(x,c)
ax.set_xlabel("x label")
ax.set_ylabel("y label")
sbnd_style.wip(ax)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import sbnd_style as sbnd_style # Applies the style automatically
from cycler import cycler

x=np.arange(0,100,1)
y=2*x
z=3*x
a=4*x
b=5*x
c=6*x

fig, ax = plt.subplots()
ax.set_prop_cycle(cycler(color=sbnd_style.SBND_COLOR_CYCLE))
ax.plot(x,y)
ax.plot(x,z)
ax.plot(x,a)
ax.plot(x,b)
ax.plot(x,c)
ax.set_xlabel("x label")
ax.set_ylabel("y label")
sbnd_style.wip(ax)
plt.show()
```
